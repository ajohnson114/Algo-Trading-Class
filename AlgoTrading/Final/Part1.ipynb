```python
def report_mean_volatility(Returns):
    """
    @param pd.DataFrame(252x30) Returns : daily excess returns of the 30 component stocks of Dow Jones Industrial
    @return pd.DataFrame(30x2) Metrics: annualized mean and volatility of the 30 component stocks
    """
    
    Metrics = pd.DataFrame(index=Returns.columns)
    
    # calculate annualized mean and volatility of daily returns
    # the annualizing factor is 252, the number of trading days in a year
    Metrics['Mean'] = np.mean(Returns)*252
    Metrics['Vol'] = np.std(Returns) * np.sqrt(252)
    
    return Metrics
```

```python
report_mean_volatility(Returns)
```
