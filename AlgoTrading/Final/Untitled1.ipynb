```python
def weight_ew(r):
    n = len(r.columns)
    ew = pd.Series(1/n, index=r.columns)
    return ew

cov = df.astype(float).cov()
cov
```

```python
def risk_weights(weights, cov):
    
    ## -- Your Code Here -- ##
    #weights.reset_index(drop=True)
    vol = np.sqrt(weights.T @ cov @ weights)
    
    risk_weights = (weights @ cov @ weights) / vol
    
    return risk_weights

risk_weights(weight_ew(df), cov)
```
