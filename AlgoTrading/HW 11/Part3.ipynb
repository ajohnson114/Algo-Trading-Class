```python
import numpy as np
import pandas as pd
import io
import statsmodels.api as sm

data_string = """date,gross,net,Total Index
4/30/1994,0.0658,0.0508,-0.0196
5/31/1994,-0.0415,-0.0315,0.0198
6/30/1994,0.1137,0.0817,-0.0104
7/31/1994,0.0355,0.0275,0.001
8/31/1994,-0.0061,-0.0051,0.0256
9/30/1994,0.0073,0.0013,0.004
10/31/1994,0.0743,0.0583,-0.0162
11/30/1994,-0.0112,-0.0082,0.0009
12/31/1994,0.0619,0.0419,-0.0052
1/31/1995,-0.0108,-0.0088,-0.0225
2/28/1995,0.0283,0.0213,0.0008
3/31/1995,0.0203,0.0133,0.0311
4/30/1995,0.0582,0.0452,0.0107
5/31/1995,0.0083,0.0063,0.0064
6/30/1995,0.0486,0.0356,-0.0007
7/31/1995,0.0308,0.0228,0.0234
8/31/1995,0.079,0.062,0.056
9/30/1995,0.0364,0.0254,-0.0008
10/31/1995,0.0306,0.0236,-0.0055
11/30/1995,0.0366,0.0286,0.026
12/31/1995,0.0833,0.0573,0.0259
1/31/1996,0.0375,0.0275,0.0652
2/29/1996,0.0093,0.0063,-0.0406
3/31/1996,0.1067,0.0767,0.0063
4/30/1996,0.0123,0.0093,0.0265
5/31/1996,0.0618,0.0488,0.0151
6/30/1996,0.0141,0.0061,0.0125
7/31/1996,-0.0073,-0.0063,-0.0456
8/31/1996,0.0671,0.0531,0.0209
9/30/1996,-0.0069,-0.0099,0.0211
10/31/1996,0.0084,0.0064,0.0289
11/30/1996,0.0308,0.0248,0.0462
12/31/1996,0.057,0.038,-0.0009
1/31/1997,0.0015,0.0005,0.0503
2/28/1997,0.0169,0.0119,0.0089
3/31/1997,0.0136,0.0056,-0.0185
4/30/1997,0.0378,0.0278,0.0244
5/31/1997,-0.0221,-0.0181,0.0047
6/30/1997,0.0134,0.0054,0.018
7/31/1997,0.0035,0.0025,0.0654
8/31/1997,0.0411,0.0311,-0.0165
9/30/1997,0.0047,-0.0013,0.0356
10/31/1997,-0.0143,-0.0123,-0.0207
11/30/1997,0.0261,0.0191,0.0051
12/31/1997,-0.0167,-0.0217,0.0285
1/31/1998,0.0077,0.0077,-0.0164
2/28/1998,-0.0071,-0.0071,0.0155
3/31/1998,0.0226,0.0146,0.055
4/30/1998,-0.0712,-0.0682,0.0053
5/31/1998,-0.1049,-0.1049,-0.0013
6/30/1998,0.0002,-0.0048,0.0111"""

def performance_part1(df):
    #  Regress the LTCM NET return (df['net']) on the Total Index (df['Total Index])
    #  Report the alpha, beta, and r-squared.
    import statsmodels.api as sm
    from sklearn.linear_model import LinearRegression
    lr = LinearRegression().fit(df[['Total Index']],df[['net']])
    alpha = float(lr.intercept_)
    beta = float(lr.coef_)
    rsquared = float(lr.score(df[['Total Index']],df[['net']]))
    # Print out alpha, beta, and r-squared
    # up to 4 digits
    print(f'alpha is {alpha:.4f}, beta is {beta:.4f}, and R^2 is {rsquared:.4f}')

def performance_part2(df):
    # Regress the LTCM NET return (df['net']) on the Total Index (df['Total Index])
    # you can use what you have in performance_part1
    #
    # Report the Information ratio of LTCM on the Total Index.
    # Store the Information ratio into a variable called <IR>
    X = sm.add_constant(df['Total Index'])
    model = sm.OLS(df['net'], X).fit()
    residual = df['net'] - model.predict(X)
    IR = model.params[0] / residual.std()

    # Print the Information ratio 
    print(f'IR is {IR:.4f}')

def test1(df):
    performance_part1(df)

def test2(df):
    performance_part2(df)

if __name__ == '__main__':
    df = pd.read_csv(io.StringIO(data_string))
    tmp = input()
    if tmp == "1":
        test1(df)
    elif tmp == "2":
        test2(df)
```
