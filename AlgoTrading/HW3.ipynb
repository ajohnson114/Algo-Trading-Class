```python
import sys


class State:
    ON = 1
    OFF = 2

class OrderState:
    OPEN = 1
    FILLED = 2
    CANCELLED = 3

class OrderSide:
    BUY = 1
    SELL = 2


class AppBase:
    def __init__(self):
        self.state = State.OFF
    def start(self):
        self.state = State.ON
    def stop(self):
        self.state = State.OFF
    def get_state(self):
        return self.state
# You need to build this class
from copy import deepcopy

class TradingStrategy(AppBase):
    def __init__(self,name):
        self.name = name
        self.order_id = 0
        self.tsorders = []
        self.state = OrderState.OPEN
        self.positions = {}
        
    def send_order(self,symbol, price, quantity, side):
        order = {"strategyname" : self.name,
                "symbol":symbol,
                'price' : price,
                'side': side,
                'quantity' : quantity,
                'order_id' : deepcopy(self.order_id),
                'state': OrderState.OPEN}
        self.order_id += 1
        self.tsorders.append(order)
        #if order['side'] ==1:
        #    self.positions[symbol] = order[quantity]
        return order
    
    def handle_order_from_market(self, order_execution):
        if order_execution['state'] == 2:
            if order_execution['side'] == OrderSide.BUY:
                self.positions[order_execution['symbol']] = order_execution['quantity']
            else:
                self.positions[order_execution['symbol']] = -order_execution['quantity']
        else:
            self.positions[order_execution['symbol']] = 0
        
            
    
    def positions(self):
        return self.positions
    
    def start(self):
        self.state = State.ON
    
    def stop(self):
        self.state = State.OFF
        
    def get_state(self):
        return self.state
    
    
    
class OrderManager(AppBase):
    def __init__(self):
        self.orders = []
        self.order_manager_id = 0
        self.positions = {}
        
    
    def handle_order_from_ts(self,order):
        if order['strategyname'] not in self.positions:
            self.positions[order['strategyname']] = {}
        order['external_order_id'] = deepcopy(self.order_manager_id)
        self.orders.append(order)
        self.order_manager_id += 1
        #print(self.positions[order['strategyname']])
        self.positions[order['strategyname']] = {order['symbol']: 0}
        #print(self.positions)
        return order
    
    def handle_order_from_market(self, order_execution):
        if order_execution['state'] == 2:
            if order_execution['side'] == OrderSide.BUY:
                self.positions[order_execution['strategyname']][order_execution['symbol']] = order_execution['quantity']
            else:
                self.positions[order_execution['strategyname']][order_execution['symbol']] = -order_execution['quantity']
                
        
        elif order_execution['state'] == 3:
            self.positions[order_execution['strategyname']][order_execution['symbol']] = 0
            self.orders.remove(order_execution)
            pass
        pass
    
    def start(self):
        self.state = State.ON
    
    def stop(self):
        self.state = State.OFF
        
    def get_state(self):
        return self.state
```

```python
def test_order_manager_market_4response(): #11
    t1 = TradingStrategy('Bob')
    t2 = TradingStrategy('Alice')
    order_sent1 = t1.send_order('MSFT', 15, 100, OrderSide.BUY)
    order_sent2 = t1.send_order('AAPL', 15, 120, OrderSide.BUY)
    order_sent3 = t2.send_order('GOOG', 10, 130, OrderSide.SELL)
    order_sent4 = t2.send_order('MSFT', 15, 140, OrderSide.SELL)

    om1 = OrderManager()
    order_sent1=om1.handle_order_from_ts(order_sent1)
    order_sent2=om1.handle_order_from_ts(order_sent2)
    order_sent3=om1.handle_order_from_ts(order_sent3)
    order_sent4=om1.handle_order_from_ts(order_sent4)

    order_sent1['state'] = OrderState.FILLED
    order_sent2['state'] = OrderState.CANCELLED
    order_sent3['state'] = OrderState.CANCELLED
    order_sent4['state'] = OrderState.FILLED

    om1.handle_order_from_market(order_sent1)
    om1.handle_order_from_market(order_sent2)
    om1.handle_order_from_market(order_sent3)
    om1.handle_order_from_market(order_sent4)
    
    print('positions', om1.positions)
    print()
    if om1.positions == {'Bob': {'MSFT': 100, 'AAPL': 0}, 'Alice': {'GOOG': 0, 'MSFT': -140}}:
        print('SUCCESS')
    else:
        print('FAIL')
```
