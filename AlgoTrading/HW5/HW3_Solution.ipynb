```python
class OrderNotFoundException(Exception):
    pass  
    
class TradingStrategy(AppBase):
    def __init__(self,name):
        self.orders = []
        self.positions = dict()
        self.name = name
        self.order_counts = 0
        
    def send_order(self, symbol, price, quantity, side):
        d = dict()
        d["strategyname"] = self.name
        d["symbol"] = symbol
        d["side"] = side
        d["price"] = price
        d["quantity"] = quantity
        d["order_id"] = self.order_counts
        self.order_counts += 1
        d["state"] = 1 # OPEN
        self.orders.append(d)
        self.positions[symbol] = 0
        return d
    
    def handle_order_from_market(self, order_execution):
        if order_execution["order_id"] > self.order_counts:
            raise OrderNotFoundException
        for i in range(len(self.orders)):
            if order_execution["order_id"] == self.orders[i]["order_id"]:
                break             
        if order_execution["state"] == 2:
            if self.orders[i]["side"] == 1:
                sign = 1
            else:
                sign = -1
            if order_execution["symbol"] in self.positions.keys():
                self.positions[order_execution["symbol"]] += order_execution["quantity"] * sign
            else:
                self.positions[order_execution["symbol"]] = order_execution["quantity"] * sign
            del self.orders[i]
        elif order_execution["state"] == 3:
            if order_execution["symbol"] not in self.positions.keys():
                self.positions[order_execution["symbol"]] = 0
            del self.orders[i]
        return order_execution
    
class OrderManager(AppBase):
    def __init__(self):
        self.external_order_counts = 0
        self.positions = dict()
        self.orders = list()
    
    def handle_order_from_ts(self,order):
        if order["strategyname"] not in self.positions.keys():
            self.positions[order["strategyname"]] = {order["symbol"]: 0}
        else:
            self.positions[order["strategyname"]][order["symbol"]] = 0
        
        d = dict()
        d["strategyname"] = order["strategyname"]
        d["symbol"] = order["symbol"]
        d["side"] = order["side"]
        d["price"] = order["price"]
        d["quantity"] = order["quantity"]
        d["order_id"] = order["order_id"]
        d["state"] = order["state"]
        d["external_order_id"] = self.external_order_counts
        self.external_order_counts += 1
        
        self.orders.append(d)
        
        # initialization
        if order["strategyname"] not in self.positions.keys():
            self.positions[order["strategyname"]] = {order["symbol"]: 0}
        else:
            self.positions[order["strategyname"]][order["symbol"]] = 0
        
        return d
    
    def handle_order_from_market(self, order_execution):
        
        if order_execution["external_order_id"] > self.external_order_counts:
            raise OrderNotFoundException
        # handle filled positions
        if order_execution["state"] == 2:
            if order_execution["side"] == 1:
                sign = 1
            else:
                sign = -1
            self.positions[order_execution["strategyname"]][order_execution["symbol"]] += order_execution["quantity"] * sign
        
        # delete filled and cancelled orders
        del self.orders[0]
        return order_execution
```
