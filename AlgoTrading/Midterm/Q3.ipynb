```python
# Step 1: Get the highest and lowest adjusted close price for the past 14 days

# Please complete the function "compute_max_min" below
# It will read a dataframe and add two columns called "max_14" and "min_14"
# "max_14" should be the highest adjusted close price for the past 14 days (inclusive)
# "min_14 should be the lowest adjusted closed price for the past 14 days (inclusive)
# If we haven't gotten price data for at least 14 days, the value should be NaN
# The function will return the dataframe

def compute_max_min(df):

    df["max_14"] = df.rolling(14).max()
    df["min_14"] = df['Adj Close'] *-1
    df["min_14"] = df['min_14'].rolling(14).max()
    df["min_14"] *= -1
    
    return df
```

```python
compute_max_min(df)
```

```python
# Step 2: Compute the stochastic oscillator

# Please complete the function "compute_stochastic_oscillator" below
# It will read the dataframe you created above and add one column called "stochastic_oscillator"
# Follow the formula in the description to compute the stochastic oscillator
# Note that the value should fall between 0 and 100
# If we haven't gotten price data for at least 14 days, the value should be NaN
# The function will return the dataframe

def compute_stochastic_oscillator(df):
    
    df["stochastic_oscillator"] = [((df["Adj Close"][i] - df["min_14"][i])/(df["max_14"][i]-df["min_14"][i]))*100 if df["max_14"][i] != np.NaN else np.NaN for i in range(len(df["Adj Close"])) ]
    return df
```

```python
compute_stochastic_oscillator(df)
```

```python
# Step 3: Identify buy & sell signals

# Please complete the function "compute_stochastic_oscillator" below
# It will read the dataframe you created above and add two columns called "buy_signal" and "sell_signal"
# buy_signal = 1 if stochastic_oscillator <= 20 otherwise 0
# sell_signal = 1 if stochastic_oscillator >= 80 otherwise 0
# If we haven't gotten price data for at least 14 days, the value should be NaN
# The function will return the dataframe

def identify_signals(df):
    
    df["buy_signal"] = [1 if df['stochastic_oscillator'][i] <= 20 else np.NaN if np.isnan(df['stochastic_oscillator'][i]) else 0 for i in range(len(df['Adj Close'])) ]
    df["sell_signal"] = [1 if df['stochastic_oscillator'][i] >= 80 else np.NaN if np.isnan(df['stochastic_oscillator'][i]) else 0 for i in range(len(df['Adj Close'])) ]
    
    return df
```

```python
def identify_signals(df):
    buy_signals = []
    sell_signals = []
    for i in range(len(df['Adj Close'])):
        if df['stochastic_oscillator'][i] == np.NaN:
            buy_signals.append(np.NaN)
            sell_signals.append(np.NaN)
        if df['stochastic_oscillator'][i] <= 20:
            buy_signals.append(1)
        else:
            buy_signals.append(0)
        if df['stochastic_oscillator'][i] >= 80:
            sell_signals.append(1)
        else:
            sell_signals.append(0)
        
    df["buy_signal"] = buy_signals
    df["sell_signal"] = sell_signals
    
    return df
```
