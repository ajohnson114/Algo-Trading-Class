```python
class ABC:
    pass
class DIRECTION:
    BUY=1
    SELL=-1
    HOLD = 0

class base_strategy(ABC):
    def predict(self):
        pass
    def fit(self,price):
        pass
```

```python


class strategy1(base_strategy):
    def __init__(self):
        self.counter = 0
        super().__init__()
    def fit(self,data):
        if len(data) == 1:
            if self.counter % 8 == 0:
                self.counter += 1
                return "buy"

            elif self.counter %4 == 0:
                self.counter += 1
                return "sell"
            else:
                self.counter += 1
        else:
            if self.counter % 8 == 0:
                self.counter += 1
                return "buy"

            elif self.counter %4 == 0:
                self.counter += 1
                return "sell"
            else:
                self.counter += 1
    
    def predict(self,price_update):
        x = self.fit(price_update)
        if x == "buy":
            return DIRECTION.BUY
        elif x == 'sell':
            return DIRECTION.SELL
        else:
            return DIRECTION.HOLD
```

```python
x = strategy1()
```
