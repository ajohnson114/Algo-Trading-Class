```python
portfolio = pd.DataFrame(np.random.rand(3366), columns = ["signal"], index = goog_data.index)
```

```python
portfolio
```

```python
portfolio['signal'] = (portfolio.signal >= 0.5).astype(int)
```

```python
portfolio
```

```python
portfolio['orders'] = portfolio['signal'].diff()
```

```python
portfolio
```

```python
# Import Matplotlib's `pyplot` module as `plt`
import matplotlib.pyplot as plt
# Plot the closing prices for `goog_data`
goog_data['Close'].plot(grid=True)
portfolio['orders'].plot()
# Show the plot
plt.show()
```
