```python
import pandas as pd
#1. Read in the meteorite data from the Meteorite_Landings.csv file:

df = pd.read_csv("Meteorite_Landings.csv")
df.head()
```

```python
#2. Rename the mass (g) column to mass, and drop all the latitude and longitude columns.

df.rename(columns={"mass (g)":"mass"}, inplace=True)

df.drop("GeoLocation", axis=1, inplace=True)
df.drop("reclat", axis=1, inplace=True)
df.drop("reclong", axis=1, inplace=True)
```
